```python

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
# Create a DataFrame to store outlier counts
outlier_summary = pd.DataFrame(columns=['Column', 'Outlier_Count', 'Percentage'])

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary = pd.concat([
        outlier_summary,
        pd.DataFrame({
            'Column': [col],
            'Outlier_Count': [outliers.shape[0]],
            'Percentage': [round(outliers.shape[0] / len(df) * 100, 2)]
        })
    ], ignore_index=True)

# Show columns with detected outliers
outlier_summary.sort_values('Outlier_Count', ascending=False)
```

```python
# Cap outliers at the 1st and 99th percentiles
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    lower_cap = df[col].quantile(0.01)
    upper_cap = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=lower_cap, upper=upper_cap)
```

```python
df['Month'] = df['Week_Date'].dt.month
df['Quarter'] = df['Week_Date'].dt.quarter
df['Year'] = df['Week_Date'].dt.year
df['WeekOfYear'] = df['Week_Date'].dt.isocalendar().week.astype(int)
```

```python
# --- Import libraries ---

import matplotlib.pyplot as plt

data = df.copy()

# --- Ensure date column is datetime type ---
data['Week_Date'] = pd.to_datetime(data['Week_Date'], errors='coerce')

# --- Create a "Month" column (Year-Month for grouping) ---
data['YearMonth'] = data['Week_Date'].dt.to_period('M')

# --- Aggregate total sales by month ---
monthly_sales = data.groupby('YearMonth')['Weekly_Sales'].sum().reset_index()

# --- Convert YearMonth back to datetime for plotting ---
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

# --- Plot line chart ---
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Weekly_Sales'], marker='o', linewidth=2)
plt.title("Total Retail Sales by Month", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# --- 0) imports
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit

# --- 1) make sure Week_Date is datetime and sort by time
df = df.copy()
df['Week_Date'] = pd.to_datetime(df['Week_Date'])
df = df.sort_values('Week_Date')

# OPTIONAL: if you created Month/Quarter/WeekOfYear earlier, keep them.
# Ensure the target and features:
target = 'Weekly_Sales'
drop_cols = ['Weekly_Sales','Week_Date']  # drop target + raw date
X = df.drop(columns=drop_cols)
y = df[target].values

# --- 2) time-based train/test split: last 20% = test
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx, :], X.iloc[split_idx:, :]
y_train, y_test = y[:split_idx], y[split_idx:]

# --- 3) fit baseline Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# --- 4) evaluate
def eval_reg(y_true, y_pred_lr, name="Model"):
    mae  = mean_absolute_error(y_true, y_pred_lr)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred_lr))
    r2   = r2_score(y_true, y_pred_lr)
    print(f"{name} -> MAE: {mae:,.2f} | RMSE: {rmse:,.2f} | R²: {r2:.3f}")
    return {"MAE":mae, "RMSE":rmse, "R2":r2}

baseline_scores = eval_reg(y_test, y_pred_lr, "Linear Regression (time split)")

# --- 5) (optional) quick residual diagnostics
residuals = y_test - y_pred_lr
print(f"Residual mean ~ {residuals.mean():.2f}, std ~ {residuals.std():.2f}")

```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

def eval_model(name, y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{name}: MAE={mae:,.2f} | RMSE={rmse:,.2f} | R²={r2:.3f}")
    return mae, rmse, r2

# --- Ridge with CV ------------------------------------------------------------
ridge = Ridge(random_state=42)
ridge_params = {'alpha': [0.01, 0.1, 1, 5, 10, 50, 100]}
ridge_cv = GridSearchCV(ridge, ridge_params, cv=5,
                        scoring='neg_root_mean_squared_error', n_jobs=-1)
ridge_cv.fit(X_train, y_train)

best_ridge = ridge_cv.best_estimator_
print(f"Best alpha (Ridge): {ridge_cv.best_params_['alpha']}")

y_pred_ridge = best_ridge.predict(X_test)
ridge_scores = eval_model("Ridge Regression", y_test, y_pred_ridge)

# --- Lasso with CV ------------------------------------------------------------
lasso = Lasso(random_state=42, max_iter=10000)
lasso_params = {'alpha': [0.001, 0.01, 0.1, 0.5, 1, 5, 10]}
lasso_cv = GridSearchCV(lasso, lasso_params, cv=5,
                        scoring='neg_root_mean_squared_error', n_jobs=-1)
lasso_cv.fit(X_train, y_train)

best_lasso = lasso_cv.best_estimator_
print(f"Best alpha (Lasso): {lasso_cv.best_params_['alpha']}")

y_pred_lasso = best_lasso.predict(X_test)
lasso_scores = eval_model("Lasso Regression", y_test, y_pred_lasso)
```

```python
ridge_coefs = pd.Series(best_ridge.coef_, index=X_train.columns).sort_values(key=abs, ascending=False)
lasso_coefs = pd.Series(best_lasso.coef_, index=X_train.columns).sort_values(key=abs, ascending=False)

print("\nTop 10 Ridge Coefficients:")
print(ridge_coefs.head(10))

print("\nTop 10 Lasso Coefficients (non-zero):")
print(lasso_coefs[lasso_coefs != 0].head(10))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def eval_model(name, y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{name}: MAE={mae:,.2f} | RMSE={rmse:,.2f} | R²={r2:.3f}")
    return mae, rmse, r2

# --- Random Forest with Grid Search ------------------------------------------
rf = RandomForestRegressor(random_state=42, n_jobs=-1)

rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

rf_cv = GridSearchCV(
    rf, rf_params, cv=3,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1, verbose=1
)
rf_cv.fit(X_train, y_train)

best_rf = rf_cv.best_estimator_
print(f"Best parameters (RF): {rf_cv.best_params_}")

y_pred_rf = best_rf.predict(X_test)
rf_scores = eval_model("Random Forest", y_test, y_pred_rf)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def eval_model(name, y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2   = r2_score(y_true, y_pred)
    print(f"{name}: MAE={mae:,.2f} | RMSE={rmse:,.2f} | R²={r2:.3f}")
    return mae, rmse, r2

# --- Neural Network base model ------------------------------------------------
mlp = MLPRegressor(random_state=42, max_iter=500)

# Hyperparameter grid (moderate size to save time)
mlp_params = {
    'hidden_layer_sizes': [(64,), (128,), (64, 32), (128, 64), (128, 128)],
    'activation': ['relu', 'tanh'],
    'alpha': [0.0001, 0.001, 0.01, 0.1],
    'learning_rate_init': [0.001, 0.01, 0.05],
    'max_iter': [300]
}

mlp_cv = GridSearchCV(
    mlp, mlp_params,
    cv=3, scoring='neg_root_mean_squared_error',
    n_jobs=-1, verbose=1
)
mlp_cv.fit(X_train, y_train)

best_mlp = mlp_cv.best_estimator_
print(f"Best parameters (MLP): {mlp_cv.best_params_}")

y_pred_mlp = best_mlp.predict(X_test)
mlp_scores = eval_model("Neural Network (MLP)", y_test, y_pred_mlp)
```
